```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Read in data
df  =  pd.read_csv('/Users/thasmias/Downloads/movies.csv')
```

```python
# Look at data
df.head()
```

```python
# Check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull()) 
    print ('{}-{}%'.format(col,pct_missing))
```

```python
# Drop missing data
df = df.dropna()
```

```python
#  Data types of columns
df.dtypes
```

```python
# change datatype
df['budget'] = df['budget'].astype('int64') # changes it to integer, removes decimal
df['gross'] = df['gross'].astype('int64') 
df['votes'] = df['votes'].astype('int64') 
df['runtime'] = df['runtime'].astype('int64') 
```

```python
df.head()
```

```python
# create YEAR column to match year released

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

df.head()
```

```python
df = df.sort_values(by = [ 'gross'], inplace = False, ascending = False ) # highest --> lowest grossing movies
```

```python
pd.set_option('display.max_rows', None )
```

```python
# drop duplicates
df ['company'].drop_duplicates().sort_values(ascending = False).head()
```

```python
df['budget'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0).astype(int)
df['gross'] = pd.to_numeric(df['gross'], errors='coerce').fillna(0).astype(int) 
```

```python
# Scatterplot of budget vs gross revenue
plt.scatter (x = df['budget'], y = df['gross'])
plt.title('Budget vs Gross Revenue')
plt.xlabel('Budget')
plt.ylabel('Gross Revenue')
plt.show ()
```

```python
df.head()
```

```python
# budget vs gross using seaborn

sns.regplot(x='budget',  y='gross', data=df, 
            scatter_kws={'color': 'orange'},  line_kws={'color':'purple'})
```

```python
# correlation 
df.corr(method='pearson',  numeric_only = True)
```

```python
# visualize corr matrix
correlation_matrix = df.corr(method ='pearson', numeric_only=True)
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel(' Movie Features ')
plt.show()
```

```python
# Company
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes 
    
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method ='pearson', numeric_only=True)
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel(' Movie Features ')
plt.show()
```

```python
correlation_matrix = df_numerized.corr()

corr_pairs = correlation_matrix.unstack()

corr_pairs.head()
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs.head()
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr

# votes & budget had the highest correlations with gross revenue
# company has low correlation with gross revenue 
```
